```python
import torch as t
import matplotlib.pyplot as plt
t.manual_seed(1)
N=100
x=t.linspace(-2,2, N)
y=3.+ x +0.2*x**2 + 0.3*t.randn(N)
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
```

```python
#fitting a straight line
ones=t.ones(N)
#feature matrix
X=t.stack([ones, x],-1) #-1 means add new dimension as the last axis
print(X.shape)
```

```python
#finding the best weights using analytical formula
w=t.linalg.inv(X.mT@X)@(X.mT@y)
print(w)
```

```python
#plotting  the prediction function
y_pred=w[0]+w[1]*x
plt.plot(x, y_pred)
plt.scatter(x,y)
```

```python
#fitting a non-linear prediction
X=t.stack([ones, x, x**2], -1)
w_2=t.linalg.inv(X.mT@X)@(X.mT@y)
y_pred2=X@w_2
plt.plot(x,y_pred2)
plt.scatter(x,y)
```

```python
#optimal weights using gradient descent
#first step, the loss function
def loss(w):
  y_pred=X@w
  return ((y-y_pred)**2).sum()
#second step is to write function for gradient of the loss
def grad(w):
  y_pred=X@w
  return -2*X.mT@(y-y_pred) #transpose replaces the sum in the loss function
#third step is to apply gradient descent
w_gd=t.zeros(3)
print(w_gd)
eta=0.001
for _ in range (100):
  w_gd=w_gd-eta*grad(w_gd)
  print(loss(w_gd))
print(w_gd)
```
